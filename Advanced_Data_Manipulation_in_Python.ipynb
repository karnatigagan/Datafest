# Joining Relational Data

Thus far we have worked with one dataset in our environment, where we loaded it in, did some visualizations or cleaning and that was it. In the real world, data scientists rarely use one csv to do analysis, and often times have to join together multiple large datasets to solve problems. There are many different ways to join dataframes, but today we will talk about three main ones:

- Left Joins
- Right Joins
- Inner Joins
- Full Outer Joins







# Joining on a key

When we join data, we want to be careful about how we go about doing it. It may be tempting to join them on some random column, but this can yield different results than we may expect, and is generally not a good practice. When data frames are joined together, you typically join them on a specific column that appears in both, called a "key". Keys are one specific column in a data frame which contain unique values, and are usually labeled as some kind of 'ID'.

This is where the word "relational" data comes from, because we join together one data frame which is "related" to another data frame based on a column they share. This key column is what we use to our advantage when specifying our joins.

To better visualize the relationships between data, we can visualize it in a "entity relationship diagram", a fancy term for an image which shows the relationship between diagrams.

# Joining Data with Pandas

Pandas offers many ways we can join together datasets, and if you look at the documentation you will notice alot of the arguments are similar. The method we will be focusing on today is the following:


```
#to merge together two data frames that share a similar key

data_frame_1.merge(data_frame_2, on='', how='', suffixes=[])

```


This method allows us to join together data very easily in one line of code. There are some arguments I want to better explain before going further:

- on (STR) ~ say two data frames share the same key to join on, we can specify this as a string to tell python which column we want to join on


- how (STR) ~ tells python the type of join we want to execute when merging the data


- suffixes (LIST) ~ more on this later, but if we join together two data frames where they both share the same column names, when joining them together we don't want them to share the same names, so we can specify how each of them should look once we have joined them. We can pass in the suffixes as a list of strings.



What if we join together two data frames with different key column names? There are some other arguments we can pass this to handle this.


```
#to merge together two data frames that share a different key

#say data_frame_1 has the key column: num_id

#say data_frame_2 has the key column: id


data_frame_1.merge(data_frame_2, left_on='', right_on='',  how='', suffixes=[])


```


The left_on and right_on arguments specify which keys we want to join on if the two data frames are different.



# Left Join

A left join returns a data frame which is all the rows in the left table, as well as the matching rows in the right data frame. It excludes any rows from the right data frame which do not match this critera. This is one of the most common methods for joining tables.


In Pandas we can use the following function to do a left join



```
# Left join df1 and df2 with a common key

df1.merge(df2, how='left', on='', suffixes=[])


#Left join df1 and df2 with two different keys

df1.merge(df2, how='left', left_on='', right_on='', suffixes=[])
```

# Right Join

A right join returns a data frame which keeps all the rows in the right data frame, as well as the matching rows from the left data frame. It excludes any observations from the left data frame that do not match the criteria. Since this mimics the results of a left join, it is less used and you will more see a left join most of the time.


To implement right join, we use the following function:



```
# Right join df1 and df2 with a common key

df1.merge(df2, how='right', on='', suffixes=[])


#Right join df1 and df2 with two different keys

df1.merge(df2, how='right', left_on='', right_on='', suffixes=[])
```

# Inner Join

An inner join returns a data frame which has matching rows from BOTH left and right data frames. It excludes any rows which do not occur in both the left AND right data frames. This is one of the more common join types, and is done quite frequently to avoid having missing values after joining tables.


To implement an inner join in Pandas, we can use the following function



```
# inner join df1 and df2 with a common key

df1.merge(df2, how='inner', on='', suffixes=[])


# inner join df1 and df2 with two different keys

df1.merge(df2, how='inner', left_on='', right_on='', suffixes=[])
```




````python
#Take the last example we just did, instead of right joining the inventory_sets_and_themes and inventories data, try an inner join

#specify the suffixes the same, and join on the same key 'set_num'

inventory_sets_themes_merged = inventory_sets_and_themes.merge(inventories, left_on='inventory_id', right_on='id', how='inner', suffixes=['_sets_and_themes', '_inventories'])

inventory_sets_themes_merged
````

#Full Outer Join

Outer joins return a data frame with rows from both tables, regardless of if it matches any criteria. It returns the rows which match in each table AND the rows which don't match in each table. Think of it as being the union of two tables.

To implement an outer join, use the following function:



```
# inner join df1 and df2 with a common key

df1.merge(df2, how='outer', on='', suffixes=[])


# inner join df1 and df2 with two different keys

df1.merge(df2, how='outer', left_on='', right_on='', suffixes=[])
```


````python
#outer join the inventory_parts data frame and part_and_partcats data frame, and specify the key as 'part_num'

#(same example as before just an outer join)

#Are there any suffixes to specify?

#save the result to a new variable: inventory_parts_partcats_outer

#How is this different from the inner join we just did?
inventory_parts_partcats_outer = inventory_parts.merge(parts_and_partcats, on='part_num', how='outer')

inventory_parts_partcats_outer
````
